```python
# Loading all data from folders. As we have files for each content.
def load_texts_labels_from_folders(path, folders):
    texts,labels = [],[]
    for idx,label in enumerate(folders):
        for fname in glob(os.path.join(path, label, '*.*')):
            texts.append(open(fname, 'r',encoding="utf8").read())
            labels.append(idx)
    # stored as np.int8 to save space 
    return texts, np.array(labels).astype(np.int8)
```

```python
#Prepare Train and Test dataset
train,train_y = load_texts_labels_from_folders(f'{IMDB_DATA_PATH}train',sentiment_classification)
val,val_y = load_texts_labels_from_folders(f'{IMDB_DATA_PATH}test',sentiment_classification)
```

```python
# Remove junk characters
re_tok = re.compile(f'([{string.punctuation}])')
def tokenize(s): return re_tok.sub(r' \1 ', s).split()
```

```python
#create term documetn matrix
#CountVectorizer converts a collection of text documents to a matrix of token counts (part of sklearn.feature_extraction.text).
vectorizer = CountVectorizer(tokenizer=tokenize)
```

```python
# fit_transform(train) finds the vocabulary in the training set. It also transforms the training set into a term-document matrix.
train_term_doc = vectorizer.fit_transform(train)
# Important: Use same vocab for validation set
val_term_doc = vectorizer.transform(val)
```

#There are 2 ways to perform sentiment analysis - 
1. Naive Bayes
2. Logistic Regression

Lets Try Naive Bayes

```python
def get_probability(y_i):
    p = x[y==y_i].sum(0)
    return (p+1) / ((y==y_i).sum()+1)
```

```python
x=train_term_doc
y=train_y

r = np.log((get_probability(1)/get_probability(0)))
b = np.log((y==1).mean() / (y==0).mean())
```

```python
r.shape,val_term_doc.shape
```

```python
# Naive Bayes Formula
pre_preds = val_term_doc @ r.T + b
preds = pre_preds.T>0
(preds==val_y).mean()
```
